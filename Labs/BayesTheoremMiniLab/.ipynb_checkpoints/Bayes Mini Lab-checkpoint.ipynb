```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

import warnings
```

```python
def cp(threshold):
    """
    This function computes the conditional probability as a function of a threshold, which can vary.
    """
    junk_long = junk[junk[:, 1] >= threshold].shape[0]
    real_long = real[real[:, 1] >= threshold].shape[0]
    return junk_long / (junk_long + real_long)
```

```python
# load data
file_location = r'D:\MATH3094\data\reviews\data.csv'
data = np.genfromtxt(file_location, delimiter=',', skip_header=1)



real = data[data[:, 0]==0, :]
print(real[:,1].mean(axis=0))
junk = data[data[:,0]==1,:]
print(junk[:,1].mean(axis=0))
```

```python
junk_long = junk[junk[:, 1] >= 100].shape[0]
junk_short = junk[junk[:, 1] < 100].shape[0]

real_long = real[real[:, 1] >= 100].shape[0]
real_short = real[real[:, 1] < 100].shape[0]

P_junk_long = junk_long / (real_long + junk_long)
print(P_junk_long)
print((junk_long + junk_short) / 5572)
```

```python
x=np.arange(2, 200)
y=np.array([cp(i) for i in x])
plt.plot(x, y, color='green')
plt.xlabel('Threshold')
plt.ylabel('Probability')
plt.title('Probability junk if longer than threshold')
```

```python
def cp2(threshold):
    """
    Probability real given short
    """
    real_short = real[real[:, 1] <= threshold].shape[0]
    junk_short = junk[junk[:, 1] <= threshold].shape[0]
    return real_short / (real_short + junk_short)
```

```python
x=np.arange(2, 200)
y=np.array([cp2(i) for i in x])
plt.plot(x, y, color='green')
plt.xlabel('Threshold')
plt.ylabel('Probability')
plt.title('Probability real if shorter than threshold')
```
